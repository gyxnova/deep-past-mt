```python
import pandas as pd
import re

subscript_map = str.maketrans("₀₁₂₃₄₅₆₇₈₉ₓ", "0123456789x")

VOWEL_MAP = {
    "á": "a2", "à": "a3",
    "é": "e2", "è": "e3",
    "í": "i2", "ì": "i3",
    "ú": "u2", "ù": "u3"
}

LETTER_MAP = {
    "Ḫ": "H", "ḫ": "h",
    "š": "sz", "Š": "SZ",
    "ṣ": "s", "Ṣ": "S",
    "ṭ": "t", "Ṭ": "T",
}

def clean_text(text):
    if pd.isna(text):
        return ""

    text = str(text)

    # Subscripts → normal
    text = text.translate(subscript_map)

        # vowels
    for k,v in VOWEL_MAP.items():
        text = text.replace(k,v)

    # consonants
    for k,v in LETTER_MAP.items():
        text = text.replace(k,v)

    # Remove line numbers like 1' 10''
    text = re.sub(r"\b\d+('{1,2})\b", "", text)


    # gaps
    text = re.sub(r"\[\s*\.{3}\s*\]", " <big_gap> ", text)
    text = re.sub(r"\[\s*x\s*\]", " <gap> ", text)
    text = text.replace("…", " <big_gap> ")

    # # protect gap tokens
    # text = text.replace("<gap>", " GAPTOKEN ")
    # text = text.replace("<big_gap>", " BIGGAPTOKEN ")

    # remove certainty + dividers
    text = re.sub(r"[!?/:]"," ",text)

    # remove half brackets
    text = text.replace("˹","").replace("˺","")

    # remove parentheses
    text = re.sub(r"\((.*?)\)","",text)

    # remove << >> and < >
    text = re.sub(r"<<(.*?)>>", r"\1", text)
    text = re.sub(r"<(?!big_gap|gap)(.*?)>", r"\1", text)

    # # restore gap tokens
    # text = text.replace("GAPTOKEN", "<gap>")
    # text = text.replace("BIGGAPTOKEN", "<big_gap>")

    # remove [] but keep inside
    text = re.sub(r"\[(.*?)\]",r"\1",text)

    # collapse spaces
    text = re.sub(r"\s+"," ",text).strip()

    return text
```

```python
train_df["clean_transliteration"] = train_df["transliteration"].apply(clean_text)
test_df["clean_transliteration"] = test_df["transliteration"].apply(clean_text)
```
