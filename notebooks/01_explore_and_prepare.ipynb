```python
import zipfile
import os

# Path to the zip in your Drive
zip_path = "/content/drive/MyDrive/deep-past/data/deep-past-initiative-machine-translation.zip"

# Path where we will extract
extract_path = "/content/drive/MyDrive/deep-past/data/extracted"

os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Extraction completed!")

```
